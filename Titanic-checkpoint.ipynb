```python
F=[]
for i in test['S'].unique():
    F.append(i)
for i in train['S'].unique():
    if i not in F:
        F.append(i)
```

```python
F
```

```python
mean_age = {}

for i in F:
    if i in test['S'].unique() and i in train['S'].unique():
        mean_age[i] = (mean_train_age[i] + mean_test_age[i])/2
    elif i in test['S'].unique():
        mean_age[i]=mean_test_age[i]
    else:
        mean_age[i]=mean_train_age[i]
```

```python
mean_age
```

```python
train['Age'] = train.apply(
    lambda row: mean_age[row['S']] if pd.isna(row['Age']) else row['Age'], axis=1
)
```

```python
test['Age'] = train.apply(
    lambda row: mean_age[row['S']] if pd.isna(row['Age']) else row['Age'], axis=1
)
```

```python
train.drop(columns=['Ticket'],inplace=True)
test.drop(columns=['Ticket'],inplace=True)
```

```python
train["Family_Members"]=train["SibSp"]+train["Parch"]+1
test["Family_Members"]=test["SibSp"]+test["Parch"]+1

```

```python
def calculate(n):
    if n==1:
        return "Alone"
    elif n>1 and n<5:
        return "Medium"
    else:
        return "Large"
```

```python
train["Family_Size"]=train["Family_Members"].apply(calculate)
```

```python
test["Family_Size"]=test["Family_Members"].apply(calculate)
```

```python
train.drop(columns=["SibSp","Parch","Family_Members","S"],inplace=True)
```

```python
train=pd.get_dummies(train,columns=['Pclass','Sex','Embarked','Family_Size'],drop_first=True)
```

```python
test=pd.get_dummies(test,columns=['Pclass','Sex','Embarked','Family_Size'],drop_first=True)
```

```python
X=train.iloc[:,1:].values
y=train.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=RandomForestClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
y_preds=clf.predict(X_test)
```

```python
precision_score(y_preds,y_test)
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV

pipe_grid = {
    "n_estimators": [100, 200, 500, 1000],                # Added 200 and 500
    "max_depth": [None, 5, 10, 20, 30],                   # Added 10, 20, and 30
    "min_samples_split": [2, 4, 6, 8, 10],                # Added 6, 8, and 10
    "max_features": ["sqrt", "log2", None]                # Added "log2" and None
}

# Create GridSearchCV object
gs_model = GridSearchCV(clf, pipe_grid, cv=5, verbose=2)

# Fit to the data
gs_model.fit(X_train, y_train)

# Output best parameters and estimator
best_params = gs_model.best_params_
best_model = gs_model.best_estimator_


```

```python
best_model = gs_model.best_estimator_
```

```python
y_preds = best_model.predict(X_test)
```

```python
precision = precision_score(y_test, y_preds, average='binary')
```
